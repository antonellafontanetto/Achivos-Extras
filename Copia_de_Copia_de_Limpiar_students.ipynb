```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/StudentsPerformance (2).csv')
#/content/estudiantes_limpio(1).csv
```

```python
df.head(10) #Observamos las primeras 10 observaciones del dataset
```

```python
df['english score'] = (df['reading score'] + df['writing score'])/2  #Creo la columna nueva english score con el promedio de reading score + writing score
```

```python
df.columns #Verifico que se haya creado la columna english score
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True) #elimino la feature Unnamed:0
```

```python
df.head()
```

```python
dummies = pd.get_dummies(df['gender']) #dummies me sirve para convertir variables categóricas en variables numéricas
df[['female', 'male']] = dummies.astype(int)
```

```python
df.head()
```

```python
df.drop('gender',axis=1,inplace=True)
```

```python
df.head()
```

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['race/ethnicity'] = le.fit_transform(df['race/ethnicity'])
df.head()
```

```python
df['lunch'].unique()
```

```python
df['race/ethnicity'].unique()
```

```python
#dummies = pd.get_dummies(df['parental level of education'])
#df[['female', 'male']] = dummies.astype(float)

from sklearn.preprocessing import OneHotEncoder
# OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, dtype=int)
onehot = ohe.fit_transform(df[['parental level of education']])
df_onehot = pd.DataFrame(onehot, columns=ohe.categories_[0])
df = pd.concat([df, df_onehot], axis=1)
df.head(15)
```

```python
df.drop('parental level of education',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# LabelEncoder
le = LabelEncoder()
df['lunch'] = le.fit_transform(df['lunch'])
df.head(10)
```

```python
#test preparation course
# LabelEncoder
le = LabelEncoder()
df['test preparation course'] = le.fit_transform(df['test preparation course'])
df.head()
```

```python
#Escalado con valores minimo y maximo determinados
from sklearn.preprocessing import MinMaxScaler



# Escalado con rango personalizado [0, 100]
scaler = MinMaxScaler(feature_range=(0, 100))

# Reemplazar directamente la columna original
df['math score'] = scaler.fit_transform(df[['math score']])

df.head()
```
